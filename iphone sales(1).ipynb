```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style = "whitegrid")
```

```python
db = pd.read_csv('apple_products (1).csv')
```

```python
print(db.describe())
```

```python
#step1: price and discount analysis 
plt.figure(figsize=(10,6))
sns.barplot(x='Product Name', y='Sale Price', data=db, palette='viridis')
plt.title('iPhone Sales Price')
plt.xticks(rotation=90)
plt.ylabel('Sale Price (INR)')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(db['Star Rating'],bins=10,kde=True)
plt.title('Distribution of star ratings')
plt.xlabel('Star Ratings')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Number Of Ratings',y = 'Number Of Reviews',hue='Star Rating',data=db,palette='cool')
plt.title('Number of Ratings & Number of Reviews')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Ram', data=db, palette='Set2')
plt.title('Distribution of RAM in iPhones')
plt.xlabel('RAM')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Step 5: Correlation between Star Rating and Sale Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sale Price', y='Star Rating', hue='Ram', data=db, palette='coolwarm')
plt.title('Sale Price vs Star Rating by RAM')
plt.xlabel('Sale Price (INR)')
plt.ylabel('Star Rating')
plt.tight_layout()
plt.show()
```
